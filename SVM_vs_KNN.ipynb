```python
import pandas as pd
import numpy as np

data = pd.read_csv('C:\Data Analyst Projects\Practice Datasets\diagnosis.csv')
```

I am going to use the same dataset as what I used for the K-Nearest-Neighbors and compare the results

```python
data.info()
```

```python
data = data.drop(['Unnamed: 32'], axis=1)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['diagnosis'], axis=1)
y = data['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import f1_score

clf = SVC(kernel='linear', C=3)

clf.fit(X_train, y_train)
```

```python
svm_accuracy = clf.score(X_test, y_test)
y_pred = clf.predict(X_test)
svm_f1 = f1_score(y_test, y_pred, pos_label='M')

print("SVM Accuracry :" + str(svm_accuracy))
print("SVM F1 Score :" + str(svm_f1))
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf2 = KNeighborsClassifier(n_neighbors=11, p=2, metric='euclidean')

clf2.fit(X_train, y_train)

y2_pred = clf2.predict(X_test)

knn_accuracy = clf2.score(X_test, y_test)
knn_f1 = f1_score(y_test, y2_pred, pos_label='M')

print("KNN Accuracry :" + str(knn_accuracy))
print("KNN F1 Score :" + str(knn_f1))
```

As we see here in this dataset, the SVM model did a better job in accuracy and F1 score, while not by a large difference, the SVM does get those last percentage points when in comparison the KNN model
