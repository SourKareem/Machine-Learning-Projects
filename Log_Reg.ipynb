You can see here that the State column, the International plan column, and the Voice mail column are objects which prevents us from scaling the data and running the model, so I will drop the State column as I don't think it will be that important, and I will be converting the Voice mail and Internation columns to boolean columns

```python
data = data.drop(['State'], axis=1)
```

```python
data['International plan'] = data['International plan'].map({'Yes': True, 'No': False})
```

```python
data['International plan'].unique()
```

```python
data['Voice mail plan'] = data['Voice mail plan'].map({'Yes': True, 'No': False})
```

```python
data['Voice mail plan'].unique()
```

```python
data.info()
```

Now that we have all the data cleaned up and ready to test and train with we will proceed with splitting into train and test 

```python
from sklearn.model_selection import train_test_split

X = data.drop(['Churn'], axis=1)
y= data['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression()

reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=reg.classes_)
disp.plot()
```

```python
print(classification_report(y_test, y_pred))
```
